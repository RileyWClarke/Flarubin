```python
import matplotlib.pyplot as plt
import numpy as np
from astropy.table import Table

file = 'cosmos2015v21.fits.gz'
tbl = Table.read(file)

stars = tbl['TYPE'] == 1

plt.figure()
plt.scatter(tbl['B_MAG_AUTO'][stars] - tbl['V_MAG_AUTO'][stars], 
            tbl['B_MAG_AUTO'][stars], s=1, alpha=0.25, c='k', label='COSMOS2015 (N='+str(np.sum(tbl['V_MAG_AUTO'][stars] <= 27))+')')

plt.gca().invert_yaxis()
plt.ylim(27,15)
plt.xlim(-0.5, 2.5)
plt.xlabel('$B-V$ [mag]')
plt.ylabel('$m_V$ [mag]')
# plt.title('COSMOS2015')
plt.plot([-0.5, 2.5], [20, 20], linestyle='--', c='C3', lw=2, alpha=0.7, label='     Gaia (N=1259)')
plt.plot([-0.5, 2.5], [24, 24], linestyle='-', c='#058B8c', lw=2, alpha=0.7, label='    Rubin (N=13430)')
plt.legend()
```
